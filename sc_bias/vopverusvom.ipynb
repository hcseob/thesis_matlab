#### Plots and outputs: Should run standalone

```python
import matplotlib.pyplot as plt
import numpy as np
import h5py
import csv
%matplotlib inline
with open('/Volumes/rboesch/cad-tsmc40/FFE_TestBenches/BiasVopAndVom_TB/MATLAB/MonteCarlo.0.csv') as csvfile:
    csvreader = csv.reader(csvfile)
    vom = []
    vop = []
    for k, row in enumerate(csvreader):
        if row and k!=0:
            vom.append(float(row[4]))
            vop.append(float(row[3]))          

plt.rc('text', usetex=True)
with h5py.File("data.hdf5", "r") as fin:
    vom_swept = fin["vom_swept"][:]
    vop_results2 = fin["vop_results2"]
    vo_results2 = fin["vo_results2"]
    plt.figure(1)
    for k in range(len(vop_results2)):
        plt.plot(vom_swept,vop_results2[k],'-k',linewidth = 0.5)
    plt.grid(True)
    plt.xlabel('$v_{om}\:[V]$')
    plt.ylabel('$v_{op}\:[V]$')
    plt.gca().set_xlim([0.5,1])
    plt.gca().set_ylim([0.1,0.6])
    plt.gcf().set_size_inches(10,6)
    plt.savefig('./figures/vopversusvom_notcalibrated.pdf')
    plt.savefig('./figures/vopversusvom_notcalibrated.png')

    plt.figure(2)
    vop_results = fin["vop_results"]
    for k in range(len(vop_results)):
        plt.plot(vom_swept,vop_results[k],'-k',linewidth = 0.5)
    plt.grid(True)
    plt.xlabel('$v_{om}\:[V]$')
    plt.ylabel('$v_{op}\:[V]$')
    plt.gca().set_xlim([0.5,1])
    plt.gca().set_ylim([0.1,0.6])
    plt.gcf().set_size_inches(10,6)
    plt.savefig('./figures/vopversusvom_calibrated.pdf')
    plt.savefig('./figures/vopversusvom_calibrated.png')
```
